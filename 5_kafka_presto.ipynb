```python
import os

PRESTO_SERVER = os.environ.get('PRESTO_SERVER')
PRESTO_USER = os.environ.get('PRESTO_USER')
```

```python
import prestodb
conn=prestodb.dbapi.connect(
    host=PRESTO_SERVER,
    port=8080,
    user=PRESTO_USER
)
cur = conn.cursor()
cur.execute('SELECT * FROM system.runtime.nodes')
rows = cur.fetchall()
rows
```

```python
import prestodb
conn=prestodb.dbapi.connect(
    host=PRESTO_SERVER,
    port=8080,
    user=PRESTO_USER,
    catalog='kafka',
    schema='default',
)
cur = conn.cursor()
cur.execute('SELECT * FROM images LIMIT 3')
rows = cur.fetchall()
rows
```
